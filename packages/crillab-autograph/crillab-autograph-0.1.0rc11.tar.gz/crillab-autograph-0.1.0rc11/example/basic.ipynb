```python
from autograph import create_plot
from autograph.core.style import LegendStyle,TextStyle
from autograph.core.enumstyle import Position,FontWeight
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 10, 100)
Y = 4+2*np.sin(2*X)
```

```python
plot = create_plot('matplotlib') # plotly is also a possible value
plot.title = "My awesome plot" # Change title

plot.title_style = TextStyle() # Title style
plot.title_style.weight=FontWeight.BOLD 
plot.title_style.font_name="Helvetica"
plot.title_style.color="blue"


plot.x_label = "X" # Change x name
plot.y_label = "Y" # Change y name

plot.x_label_style=TextStyle()
plot.x_label_style.weight=FontWeight.N_900
plot.x_label_style.color="red"
plot.x_label_style.font_name="Latin Modern Mono Prop"

plot.plot(X,Y,label="label")


plot.legend=LegendStyle()
plot.legend.position=Position.RIGHT
plot.legend.title="title"

plot.show()                                                     
```
