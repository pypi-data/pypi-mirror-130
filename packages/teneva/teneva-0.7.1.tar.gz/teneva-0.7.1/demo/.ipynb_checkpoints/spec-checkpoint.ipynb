```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import teneva
from time import perf_counter as tpc
np.random.seed(42)
```

```python
data = []
with open('./INTERNAL_temperature_dependent_activity_coefficients.csv', newline='') as f:
    for k, row in enumerate(csv.reader(f, delimiter=',')):
        if k == 0:
            continue
        data.append([int(row[0]), int(row[1]), float(row[2]), float(row[3])])

# Solvent to zero:
for d in data:
    d[0] = int(d[0] - 1)
    
# Solute to zero:
for d in data:
    d[1] = int(d[1] - 1)
    
# Temperature to index:
for d in data:
    d[3] = int(d[3] - 126.15)
```

```python
I = [d[0] for d in data]
J = [d[1] for d in data]
A = [d[2] for d in data]
T = [d[3] for d in data]

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(14, 14))
plt.subplots_adjust(wspace=0.2, hspace=0.2)

ax1.hist(I, bins=50)
ax1.semilogy()
ax1.set_xlabel('Solvent')
ax1.set_title(f'(min: {np.min(I)}; max: {np.max(I)}; different values:{len(list(set(I)))})')

ax2.hist(J, bins=50)
ax2.semilogy()
ax2.set_xlabel('Solute')
ax2.set_title(f'(min: {np.min(J)}; max: {np.max(J)}; different values:{len(list(set(J)))})')

ax3.hist(T, bins=50)
ax3.semilogy()
ax3.set_xlabel('Temperature (index)')
ax3.set_title(f'(min: {np.min(T)}; max: {np.max(T)}; different values:{len(list(set(T)))})')

ax4.hist(A, bins=50)
ax4.semilogy()
ax4.set_xlabel('Activity coefficient')
ax4.set_title(f'(min: {np.min(A):-.4f}; max: {np.max(A):-.4f}; different values:{len(list(set(A)))})')

plt.show()
```

```python
K_trn = 10000
K_tst = len(data) - K_trn
print(K_tst)
```

```python
ind = np.arange(len(data))
np.random.shuffle(ind)
ind_trn = ind[:K_trn]
ind_tst = ind[K_trn:]
```

```python
data_trn = [data[i] for i in ind_trn]
I_trn = np.array([[d[0], d[1], d[3]] for d in data_trn])
Y_trn = np.array([d[2] for d in data_trn])
```

```python
data_tst = [data[i] for i in ind_tst]
I_tst = np.array([[d[0], d[1], d[3]] for d in data_tst])
Y_tst = np.array([d[2] for d in data_tst])
```

```python
d = 3
N = [65115, 28350, 451]
```

```python
t = tpc()
Y = teneva.ANOVA(I_trn, Y_trn, order=1)
t = tpc() - t

print(f'Build time     : {t:-10.2f}')
```

```python
Z = Y(I_trn)
e_trn = np.linalg.norm(Z - Y_trn) / np.linalg.norm(Y_trn)

Z = Y(I_tst)
e_tst = np.linalg.norm(Z - Y_tst) / np.linalg.norm(Y_tst)

print(f'Error on train : {e_trn:-10.2e}')
print(f'Error on test  : {e_tst:-10.2e}')
```

```python
get = teneva.getter(Y)

Z = np.array([get(i) for i in I_trn])
e_trn = np.linalg.norm(Z - Y_trn) / np.linalg.norm(Y_trn)

Z = np.array([get(i) for i in I_tst])
e_tst = np.linalg.norm(Z - Y_tst) / np.linalg.norm(Y_tst)

print(f'Error on train : {e_trn:-10.2e}')
print(f'Error on test  : {e_tst:-10.2e}')
```
